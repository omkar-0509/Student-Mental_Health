### Matplotlib Functions

```python
df_sample = df.sample(1000)

plt.plot(df_sample['stress_level'])
plt.show()
```

```python
df.groupby('gender')['stress_level'].mean().plot(kind='bar')
plt.title("Gender vs Stress Level")
plt.show()
```

```python
plt.hist(df['stress_level'])
plt.title("Stress Level Distribution")
plt.show()
```

```python
plt.scatter(df['anxiety_score'], df['depression_score'])
plt.xlabel("Anxiety Score")
plt.ylabel("Depression Score")
plt.title("Anxiety vs Depression")
plt.show()
```

```python
plt.boxplot(df['burnout_score'])
plt.title("Burnout Score Boxplot")
plt.show()
```

```python
df['risk_level'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Risk Level Distribution")
plt.show()
```

```python
plt.pie(df['risk_level'].value_counts(), labels=df['risk_level'].unique())
plt.show()
```

```python
plt.plot(df['stress_level'], label='Stress')
plt.plot(df['anxiety_score'], label='Anxiety')
plt.plot(df['depression_score'], label='Depression')

plt.legend()
plt.title("Mental Health Comparison")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.hist(df['stress_level'])

plt.subplot(2,2,2)
plt.hist(df['anxiety_score'])

plt.subplot(2,2,3)
plt.hist(df['depression_score'])

plt.show()
```
